```python
import os
import cv2

# Paths
image_dir = '/content/gdrive/MyDrive/content/vdvwc_data/train/images'
label_dir = '/content/gdrive/MyDrive/content/vdvwc_data/train/labels'
output_image_dir = '/content/gdrive/MyDrive/vdvwcresizeddata512/train/images/'
output_label_dir = '/content/gdrive/MyDrive/vdvwcresizeddata512/train/labels/'

# Desired new size
new_width, new_height = 512, 512

# Ensure output directories exist
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)

# Function to resize images and adjust labels
def resize_image_and_labels(image_path, label_path, new_width, new_height):
    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Cannot load image: {image_path}")
        return

    # Get original dimensions
    original_height, original_width = img.shape[:2]

    # Resize image
    resized_img = cv2.resize(img, (new_width, new_height))

    # Read and adjust labels
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()

        new_labels = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) == 5:
                class_id = parts[0]
                x_center = float(parts[1]) * original_width
                y_center = float(parts[2]) * original_height
                width = float(parts[3]) * original_width
                height = float(parts[4]) * original_height

                # Adjust for new image size
                new_x_center = (x_center / original_width) * new_width / new_width
                new_y_center = (y_center / original_height) * new_height / new_height
                new_width = (width / original_width) * new_width / new_width
                new_height = (height / original_height) * new_height / new_height

                new_labels.append(f"{class_id} {new_x_center:.6f} {new_y_center:.6f} {new_width:.6f} {new_height:.6f}")

        # Save adjusted labels
        with open(os.path.join(output_label_dir, os.path.basename(label_path)), 'w') as f:
            for label in new_labels:
                f.write(label + '\n')

    # Save resized image
    cv2.imwrite(os.path.join(output_image_dir, os.path.basename(image_path)), resized_img)

# Process all images and labels
for img_file in os.listdir(image_dir):
    if img_file.endswith('.jpg') or img_file.endswith('.png'):
        img_path = os.path.join(image_dir, img_file)
        label_path = os.path.join(label_dir, img_file.replace('.jpg', '.txt').replace('.png', '.txt'))
        resize_image_and_labels(img_path, label_path, new_width, new_height)

print("Image resizing and label adjustment complete.")
```

```python
import os
import cv2

# Paths
image_dir = '/content/gdrive/MyDrive/content/vdvwc_data/val/images'
label_dir = '/content/gdrive/MyDrive/content/vdvwc_data/val/labels'
output_image_dir = '/content/gdrive/MyDrive/vdvwcresizeddata512/val/images/'
output_label_dir = '/content/gdrive/MyDrive/vdvwcresizeddata512/val/labels/'

# Desired new size
new_width, new_height = 512, 512

# Ensure output directories exist
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)

# Function to resize images and adjust labels
def resize_image_and_labels(image_path, label_path, new_width, new_height):
    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Cannot load image: {image_path}")
        return

    # Get original dimensions
    original_height, original_width = img.shape[:2]

    # Resize image
    resized_img = cv2.resize(img, (new_width, new_height))

    # Read and adjust labels
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()

        new_labels = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) == 5:
                class_id = parts[0]
                x_center = float(parts[1]) * original_width
                y_center = float(parts[2]) * original_height
                width = float(parts[3]) * original_width
                height = float(parts[4]) * original_height

                # Adjust for new image size
                new_x_center = (x_center / original_width) * new_width / new_width
                new_y_center = (y_center / original_height) * new_height / new_height
                new_width = (width / original_width) * new_width / new_width
                new_height = (height / original_height) * new_height / new_height

                new_labels.append(f"{class_id} {new_x_center:.6f} {new_y_center:.6f} {new_width:.6f} {new_height:.6f}")

        # Save adjusted labels
        with open(os.path.join(output_label_dir, os.path.basename(label_path)), 'w') as f:
            for label in new_labels:
                f.write(label + '\n')

    # Save resized image
    cv2.imwrite(os.path.join(output_image_dir, os.path.basename(image_path)), resized_img)

# Process all images and labels
for img_file in os.listdir(image_dir):
    if img_file.endswith('.jpg') or img_file.endswith('.png') or img_file.endswith('.jpeg'):
        img_path = os.path.join(image_dir, img_file)
        label_path = os.path.join(label_dir, img_file.replace('.jpg', '.txt').replace('.png', '.txt'))
        resize_image_and_labels(img_path, label_path, new_width, new_height)

print("Image resizing and label adjustment complete.")
```
